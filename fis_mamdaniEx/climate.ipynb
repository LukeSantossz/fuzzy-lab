```python
import numpy as np 
import skfuzzy as fuzz 
from skfuzzy import control as ctrl
```

```python
# Antecedente e Consequente // Input e Output

# Antecedentes

# Intervalos (np.arange(0,0,0) precisam de confirmação baseado na climatologia registrada da região e literatura técnica.
temperature = ctrl.Antecedent(np.arange(0,60,0.1), 'Temperatura')
humidity = ctrl.Antecedent(np.arange(0,101,1.0), 'Umidade')
rain = ctrl.Antecedent(np.arange(0,500,0.2), 'Chuva')
wind = ctrl.Antecedent(np.arange(0,150,0.5), 'Vento')
dt = ctrl.Antecedent(np.arange(0,40,0.1), 'Delta T')
```

```python
# Consequentes

# Mapear Todos os Consequentes conforme intervalo:
spray_recomendation = ctrl.Consequent(np.arange(0, 11, 1),'sp')

spray_recomendation['proibida'] = fuzz.trimf(spray_recomendation.universe, [0, 0, 4])
spray_recomendation['atencao'] = fuzz.trimf(spray_recomendation.universe, [3, 5, 7])
spray_recomendation['janela_disponivel'] = fuzz.trimf(spray_recomendation.universe, [6, 10, 10])

water_stress = ctrl.Consequent(np.arange(0, 1.1, 0.1),'eh')
irr_recomendation = ctrl.Consequent(np.arange(0, 11, 1),'ri')

# Essa aqui é do capeta -- Precisamos verificar certinho a literatura técnica da área para confirmar os valores.
bet_productivity = ctrl.Consequent(np.arange(0, 101, 1),'pd')
```

```python
# Funções de Pertinência

temperature.automf(number=7, names=['frio_extremo','frio','ameno','ideal','quente','estresse_termico','critico'])
humidity.automf(number=7,names=['deserto','critico_seco','seco','ideal','umido','saturado','condensacao'])
rain.automf(number=7,names=['seco','trace','leve','moderada','forte','muito_forte','extrema'])
wind.automf(number=7,names=['calmo','ideal_pulv','moderado','atencao_pulv','forte','muito_forte','tempestade'])
dt.automf(number=7,names=['inversao_termica','ideal_pulv','aceitavel','atencao','critico_pulv','proibido','extremo'])
```

```python
# Regras de Pertinência

# Mapear Todas as Regras conforme intervalos e literaturas:

# Regra Pulverização com janela_disponivel (Faltam: atencao e  proibida)!
r1 = ctrl.Rule(
    dt['ideal_pulv'] &
    (rain['seco'] | rain['trace']) &
    humidity['ideal'] &
    temperature['ideal'],
    spray_recomendation['janela_disponivel']
)
```
